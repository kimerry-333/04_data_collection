## Connector
- db 연결 정보를 작성
- 작성된 정보를 통해서 실제로 DB 서버와 연결하는 객체

## Cursor
- SQL을 수행하고, 결과를 반환 받는 객체

**두 객체 모두 사용 후 반납(close) 처리 필수**

```python
import mysql.connector

# DB 연결 정보 작성
config = {
    'host' : 'localhost',   #서버주소
    'port' : 3306,          #기본 값이 3306(3306인 경우 생략 가능)
    'user' : 'skn_ai',
    'password' : '1234',
    'database' : 'menudb'
}

# 커넥터 객체 생성
conn = mysql.connector.connect(**config)    # ** : dict unpacking

# 커서 객체 생성
cursor = conn.cursor()

# select 수행
# [주의] SQL 마지막에 세미콜론(;) 작성X
cursor.execute('select * from tbl_menu')
rows = cursor.fetchall()    #결과집합(ResultSet) 반환
for row in rows:
    print(row)

# 사용 완료된 db 객체 close(생성 역순으로 close)
cursor.close()
conn.close()
```

```python
# 한 행을 표현하기 위한 Menu 클래스 정의
class Menu:
    def __init__(self,menu_code,menu_name,menu_price,category_code,orderable_status):
        self.__menu_code = menu_code
        self.__menu_name = menu_name
        self.__menu_price = menu_price
        self.__category_code = category_code
        self.__orderable_status = orderable_status

    # 프로퍼티 데코레이터
    # - getter 역할을 하는 메서드
    # - 외부 호출 시 () 생략 가능 -> menu.menu_code
    @property
    def menu_code(self):
        return self.__menu_code

    @property
    def menu_name(self):
        return self.__menu_name

    @property
    def menu_price(self):
        return self.__menu_price

    @property
    def category_code(self):
        return self.__category_code

    @property
    def orderable_status(self):
        return self.__orderable_status

    # __repr__ 메서드 : 객체에 어떤 값이 저장되어있는지 하나의 문자열로 반환
    # repr == representation(표현)
    def __repr__(self):
        return f"Menu(menu_code={self.__menu_code},menu_name={self.__menu_name},menu_price={self.__menu_price},category_code={self.__category_code},orderable_status={self.__orderable_status})"
```

```python
# 메뉴 테이블 전체 조회에서 Menu 리스트애 담기
try:
    with mysql.connector.connect(**config) as conn:  # DB 연결 담당
        with conn.cursor() as cursor:                # SQL 수행 담당

            cursor.execute('select * from tbl_menu') # 실행

            rows=cursor.fetchall()                   # 조회 결화 집합 반환

            # 조회 결과를 Menu 객체 list로 반환
            menus = [Menu(*row) for row in rows]

    for menu in menus:
        print(menu)

except mysql.connector.Error as err:
    print("DB 오류 : ", err)
```

```python
# 메뉴 테이블 메뉴명에 입력된 단어가 포함된 메뉴만 조회
input_menu_name = input("메뉴명 포함 단어 입력")

try:
    with mysql.connector.connect(**config) as conn:  # DB 연결 담당
        with conn.cursor() as cursor:                # SQL 수행 담당

            cursor.execute(f"""
                           select *
                           from tbl_menu
                           where menu_name like '%{input_menu_name}%'
                           """) # 실행

            rows=cursor.fetchall()                   # 조회 결화 집합 반환

            # 조회 결과를 Menu 객체 list로 반환
            menus = [Menu(*row) for row in rows]

    for menu in menus:
        print(menu)

except mysql.connector.Error as err:
    print("DB 오류 : ", err)
```

### Record 추가(insert)

```python
new_menu = Menu(None,'멸치액젓맛요거트',20000,8,'Y')

try:
    with mysql.connector.connect(**config) as conn:
        with conn.cursor() as cursor:
            cursor.execute('''
                insert into tbl_menu(
                    menu_name,
                    menu_price,
                    category_code,
                    orderable_status
                    )
                values(%s,%s,%s,%s)
            ''', (
                new_menu.menu_name,
                new_menu.menu_price,
                new_menu.category_code,
                new_menu.orderable_status
            ))
            # %s : mysql connect에서 빈칸 하나
            # %s 자리 대입 시 자동으로 양쪽에 '' 작성

            #반영된 행의 수 확인
            print("insert된 행의 수 : ",cursor.rowcount)

            #커밋
            conn.commit()

except mysql.connector.Error as err:
    print('DB에러 : ',err)

# insert된 메뉴 pk 확인 (103)
```

## 레코드 수정(update)

```python
# 추가한 메뉴의 가격을 10000 인하
target_menu_code = 103

try:
    with mysql.connector.connect(**config) as conn:
        with conn.cursor() as cursor:

            cursor.execute('''
                update tbl_menu
                set menu_price = menu_price - 10000
                where menu_code = %s
            ''', (target_menu_code, ))
                # 튜플 타입으로 전달되어야하는데
                # target_menu_code만 전달할 경우 int 타입으로 전달됨
                # -> (target_menu_code, ) : 튜플타입

            print("수정된 행의 수: ", cursor.rowcount)

            if cursor.rowcount > 0:
                conn.commit()
            # (참고) 수정된 행이 있을 때만 커밋

except mysql.connector.Error as err:
    print('DB에러 : ',err)
```

## 레코드 삭제(delete)

```python
# 추가한 메뉴 삭제
try:
    with mysql.connector.connect(**config) as conn:
        with conn.cursor() as cursor:

            cursor.execute('''
                delete from tbl_menu
                where menu_code = %s
            ''',(target_menu_code,))

            print("수정된 행의 수: ", cursor.rowcount)

            if cursor.rowcount > 0:
                conn.commit()

except mysql.connector.Error as err:
    print('DB에러 : ',err)

```
